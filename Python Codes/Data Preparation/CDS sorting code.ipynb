```python
new_row = [dictionary2[col] for col in df1.columns]

# Insert the new row at the first position in df1
df1.loc[-1] = new_row  # Add new row at the end
df1.index = df1.index + 1  # Shift the index
df1 = df1.sort_index() 


df1_transposed = df1.transpose()
df1_transposed_sorted = df1_transposed.sort_values(by=0)
df1 = df1_transposed_sorted.transpose()
```

```python
df_columns = set(df1.columns)
dict_keys = set(dictionary2.keys())

# Find columns in DataFrame that are not in the dictionary
missing_in_dict = df_columns - dict_keys

# Find keys in the dictionary that are not in the DataFrame columns
extra_in_dict = dict_keys - df_columns

print("Columns in DataFrame not in dictionary:", missing_in_dict)
print("Keys in dictionary not in DataFrame columns:", extra_in_dict)
```
