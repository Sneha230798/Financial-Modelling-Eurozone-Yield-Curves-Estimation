```python
import pandas as pd
import numpy as np
```

```python
country = 'Greece'
```

```python
df = pd.read_excel(f"C:/Users/defaultuser0/Desktop/Sneha_Thesis/excel_files/Sov_Coupons/{country}_sov_coupon.xlsx")
```

```python
def replace_non_na(row):
    return row.apply(lambda x: row['Code'] if pd.notna(x) else x)

# Apply the function to each row
df= df.apply(replace_non_na, axis=1)

df = df.drop('Code', axis = 1)

dfs = [pd.DataFrame({col: df[col].dropna()}).reset_index(drop=True) for col in df.columns]

# Combine these DataFrames with an outer join
result = pd.concat(dfs, axis=1, join='outer')

# Reset index for cleaner output
result = result.reset_index(drop=True)

df = result.copy()
```
